## **What is the average price of cars by fuel type?**

```python
# Calculate average price by fuel type
avg_price = df.groupby('Fuel type')['Price'].mean().sort_values()

# Plot
plt.figure(figsize=(10, 6))
avg_price.plot(kind='bar', color='skyblue')
plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price ()')
plt.xticks(rotation=45)
plt.show()
```

# **How does car price correlate with manufacturing year?**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Manufacturing_year'], df['Price'], alpha=0.5)
plt.title('Price vs. Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Price ()')
plt.grid(True)
plt.show()
```

## **What is the distribution of car prices?**

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=30, color='green', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price ()')
plt.ylabel('Frequency')
plt.show()
```

## **Which transmission type is more common?**

```python
transmission_counts = df['Transmission'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(transmission_counts, labels=transmission_counts.index, autopct='%1.1f%%')
plt.title('Transmission Type Distribution')
plt.show()
```

# **Do cars with a spare key have higher prices?**

```python
spare_key_price = df.groupby('Spare key')['Price'].mean()
print(spare_key_price)
```

# **Top 5 most common car models**

```python
top_models = df['Model Name'].value_counts().head(5)
top_models.plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('Top 5 Most Common Models')
plt.xlabel('Model')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

# **How does engine capacity affect price?**

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine capacity'], df['Price'], alpha=0.5)
plt.title('Engine Capacity vs. Price')
plt.xlabel('Engine Capacity (cc)')
plt.ylabel('Price ()')
plt.show()
```

## **Average kilometers driven by ownership count**

```python
df.groupby('Ownership')['KM driven'].mean().plot(kind='bar')
plt.title('Average KM Driven by Ownership')
plt.xlabel('Ownership Count')
plt.ylabel('KM Driven')
plt.show()
```

# **Price distribution by transmission type**

```python
sns.boxplot(x='Transmission', y='Price', data=df)
plt.title('Price Distribution by Transmission')
plt.show()
```

# **Are imperfections related to price?**

```python
sns.scatterplot(x='Imperfections', y='Price', data=df)
plt.title('Imperfections vs. Price')
plt.show()
```

# **Most common fuel types**

```python
df['Fuel type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Fuel Type Distribution')
plt.show()
```

## **Yearly trend of car listings**

```python
yearly_counts = df['Manufacturing_year'].value_counts().sort_index()
yearly_counts.plot(kind='line', marker='o')
plt.title('Car Listings by Manufacturing Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

### **Do repainted parts affect price?**

```python
sns.scatterplot(x='Repainted Parts', y='Price', data=df)
plt.title('Repainted Parts vs. Price')
plt.show()
```

# **Kilometers driven vs. manufacturing year**

```python
sns.scatterplot(x='Manufacturing_year', y='KM driven', data=df)
plt.title('KM Driven vs. Manufacturing Year')
plt.show()
```

# **Fuel type vs. transmission type**

```python
pd.crosstab(df['Fuel type'], df['Transmission']).plot(kind='bar')
plt.title('Fuel Type vs. Transmission')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()
```

##  **Price distribution for CNG vs. Petrol cars**

```python
sns.kdeplot(data=df, x='Price', hue='Fuel type', fill=True)
plt.title('Price Distribution by Fuel Type')
plt.show()
```

## **Top 5 models with the highest mileage**

```python
top_mileage = df.nlargest(5, 'KM driven')[['Model Name', 'KM driven']]
print(top_mileage)
```
