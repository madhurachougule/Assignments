Which industry has the highest job loss due to AI?

```python
data = {
    'Industry': ['Gaming', 'Media', 'Automotive', 'Legal'],
    'Job Loss (%)': [18, 17, 11, 47],
}
df = pd.DataFrame(data)

plt.figure(figsize=(9, 6))
sns.set_style("whitegrid")

# Bar plot
ax = sns.barplot(x='Industry', y='Job Loss (%)', data=df, palette="viridis")

# Annotations
plt.title(" Industriwise Job Loss Due to AI", pad=20, fontweight='bold')
plt.xlabel("Industry", labelpad=10)
plt.ylabel("Percentage (%)", labelpad=10)
plt.ylim(0, 50)

# Add percentage labels
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%",
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

**Which country has the highest job loss rate due to AI?**

```python
countries = ['USA', 'China', 'South Korea', 'France']
job_loss_rates = [11, 47, 17, 28]

plt.figure(figsize=(10, 6))
plt.bar(countries, job_loss_rates, color='#ff9999', edgecolor='black', width=0.6)
plt.title('Job Loss Rates Due to AI', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Job Loss Rate (%)', fontsize=12)
plt.ylim(5, 50)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, rate in enumerate(job_loss_rates):
    plt.text(i, rate + 0.5, f'{rate}%', ha='center', fontsize=10)

plt.show()
```
